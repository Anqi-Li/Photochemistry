```python
# FIRST FUNTION IS THE O2 VOLUME FOR THE SCHUMANN RUNGE BANDS PROGRAM
#
#
# SECOND FUNTION IS THE PHOTOLYSIS PROGRAM FOR THE PHOTOLYSIS RATES FOR EACH REGION 
# AS WELL AS FOR THE O2 AND O3 PHOTOLYSIS RATES

def O2_vol(z):
    #z in km
    n0 = popt[0]
    H = popt[1]
    return H * 1e5 * n0 * (np.exp(-z/H)-np.exp(-105/H))

def photolysis(z, T, sol_zen, o, o2, o3, n2, co2):
    #z in m
    #o2, o3 in cm-3
    if sol_zen == 0 or sol_zen == 180:
        sol_zen += 0.001
    def path_z(z_top, z_t, sol_zen, nsteps):
        Re = 6375e3 #m
        sol_zen /= 180/np.pi
        B = np.arcsin((Re+z_t) * np.sin(np.pi-sol_zen)/(Re+z_top))
        S_top = np.sin(sol_zen-B)*(Re+z_top)/np.sin(np.pi-sol_zen)
        
        Ret2 = (Re+z_t)**2
        step = S_top/nsteps
        S_top_half = S_top - step/2
        z_step = [np.sqrt(Ret2 + (S_top_half - i*step)**2 - 
                         2*(Re + z_t)*(S_top_half - 
                                      i*step)*np.cos(np.pi-sol_zen))-Re for i in range(nsteps)]
        return (z_step,step)
    
    
    Jsrb = []
    Jsrc = []
    Jlya = []
    Jherz= []
    Jhart= []
    J1 = []
    J2 = []
    J3 = []
    J4 = []
    for iz, z_t in enumerate(z):
        
        so2[65:82] = effxsect(T[iz], [O2_vol(z_t*1e-3)])
        z_paths, path_step = path_z(z[-1],z_t,sol_zen, 500)
        #print(np.array(z_paths)/1e3)
        tau = (so2*(np.exp(np.interp(z_paths,z,np.log(o2)))).sum()+
               so*(np.interp(z_paths,z,o)).sum()+
               sn2*(np.interp(z_paths,z,n2)).sum()+
               sco2*(np.interp(z_paths,z,co2)).sum()+
               so3*(np.interp(z_paths,z,o3)).sum()) * path_step * 1e2 #m->cm
        
        jo2, jo3 = (irrad * (so2, so3) * np.exp(-tau))
        
        Jhart.append(jo3[np.transpose(np.logical_and(wave>210,wave<310))].sum())
        Jsrc.append(jo2[np.transpose(np.logical_and(wave>122,wave<175))].sum())
        Jsrb.append(jo2[np.transpose(np.logical_and(wave>175,wave<200))].sum())
        Jherz.append(jo2[np.transpose(np.logical_and(wave>194,wave<240))].sum())
        Jlya.append(jo2[np.transpose(wave==121.567)].sum())
        J1.append(jo2[np.transpose(np.logical_and(wave>177.5,wave<256))].sum())
        J2.append(jo2[np.transpose(np.logical_and(wave<177.5,wave>0))].sum())
        J3.append(jo3[np.transpose(np.logical_and(wave>200.0,wave<320.0))].sum()*0.1+jo3[np.transpose(np.logical_and(wave>320.0,wave<730.0))].sum())
        J4.append(jo3[np.transpose(np.logical_and(wave>167.5,wave<200.0))].sum()+jo3[np.transpose(np.logical_and(wave>200.0,wave<320.0))].sum()*0.9)

        
    return np.array(Jhart), np.array(Jherz), np.array(Jsrc), np.array(Jsrb), np.array(Jlya), np.array(J1), np.array(J2), np.array(J3), np.array(J4)
```

```python
# CREATION OF THE INTERPOLATED gA
# AND ALSO THE INTERPOLATED O2 & O3 PHOTOLYSIS RATES
# FOR EACH ALTITUDE FOR THE SIMULATION PROGRAM

def create_G(h,t):
    def function():
        string = 'G_factor_all' + "[" + str(h) + ",:]"
        return interp1d(t_final,eval(string + ".tolist()"))(t)
    return function

def create_J1(h,t):
    def function():
        string = 'J1_all' + "[" + str(h) + ",:]"
        return interp1d(t_final,eval(string + ".tolist()"))(t)
    return function

def create_J2(h,t):
    def function():
        string = 'J2_all' + "[" + str(h) + ",:]"
        return interp1d(t_final,eval(string + ".tolist()"))(t)
    return function

def create_J4(h,t):
    def function():
        string = 'J4_all' + "[" + str(h) + ",:]"
        return interp1d(t_final,eval(string + ".tolist()"))(t)
    return function

def norm_t(t):
        return np.mod(t,24*60*60) 
```

```python
# SAME AS BEFORE BUT THIS IS FOR THE SECOND ITERATION 
# WITH THE PHOTOLYSIS RATES BEING RE-EVALUATED 
# AFTER THE FIRST SIMULATION

def create_new_G(h,t):
    def function():
        string = 'G_factor_all' + "[" + str(h) + ",:]"
        return interp1d(t_final,eval(string + ".tolist()"))(t)
    return function

def create_new_J1(h,t):
    def function():
        string = 'new_J1_all' + "[" + str(h) + ",:]"
        return interp1d(t_final,eval(string + ".tolist()"))(t)
    return function

def create_new_J2(h,t):
    def function():
        string = 'new_J2_all' + "[" + str(h) + ",:]"
        return interp1d(t_final,eval(string + ".tolist()"))(t)
    return function

def create_new_J4(h,t):
    def function():
        string = 'new_J4_all' + "[" + str(h) + ",:]"
        return interp1d(t_final,eval(string + ".tolist()"))(t)
    return function

def norm_t(t):
        return np.mod(t,24*60*60) 
```

```python
# CREATING A FUNCTION THAT CREATES A STRING 
# THAT IS GOING TO BE COMPLETED WITH THE PROPER 
# ALTITUDE IN THE SIMULATION PROGRAM

def rates():
    string = ['create_J1', 'create_J2', 'create_J4', 'create_G']
    return string

def new_rates():
    string = ['create_new_J1', 'create_new_J2', 'create_new_J4', 'create_new_G']
    return string
```

```python
# SIMULATION PROGRAM THAT CONSIST IN AN ODEINT INTEGRATION
# FOR EACH OF THE ALTITUDES 
# STARTING AT MIDDAY (12:00) OF THE 1ST DAY AND 
# FINISHING AT MIDNITH (00:00) OF THE 4RD DAY 
# FOR A TOTAL OF 3 DAYS AND A HALF OF INTEGRATION
# WITH A TIME STEP OF 4 SECONDS
#
#
# RATES TO BE UPDATED TO ANQI'S

def simulation(z):

    x0 = [n_o3[z],
          n_o2[z],
          n_o2_1Delta[z],
          n_o2_1Sigma[z],
          n_o[z],
          n_o_1D[z],
          n_oh[z],
          n_ho2[z],
          n_h[z]]
    
    photolysis_rates = rates()
    print(f'z = {5*z+50}')
    
    def system(values, t):
        #First the concentrations
        
        O3        = values[0]     
        O2        = values[1]     
        O2_1Delta = values[2]
        O2_1Sigma = values[3]    
        O         = values[4] 
        O_1D      = values[5]       
        OH        = values[6]      
        HO2       = values[7]      
        H         = values[8]         
        N2        = 4.0*O2
        M         = O2 + N2
        
        #O3 reactions & rates
        
        J4 = eval(photolysis_rates[2] + "(z,norm_t(t))()")
        k2 = 1.0e-10*np.exp(-516/T[z])
        k3 = 1.5e-12*np.exp(-1000/T[z])
        
        r1 = J4 * O3 
        r2 = k2 * O3 * H 
        r3 = k3 * O3 * OH
        
        #O2 reactions
        
        k4 = 1.07e-34*np.exp(510/T[z])
        k5 = 2.08e-32*np.exp(290/T[z])
        J1 = eval(photolysis_rates[0] + "(z,norm_t(t))()")
        J2 = eval(photolysis_rates[1] + "(z,norm_t(t))()")
        G  = eval(photolysis_rates[3] + "(z,norm_t(t))()")
        
        r4 = k4 * O2 * O * M
        r5 = k5 * O2 * H * M
        r6 = J1 * O2
        r7 = J2 * O2
        r20 = G * O2
        
        #O2_1Delta reactions
        
        k8  = 3.6e-18*np.exp(-220/T[z])
        k9  = 1.0e-20
        k10 = 1.3e-16
        A1 = 2.23e-4
        
        r8  =  k8 * O2_1Delta * O2
        r9  =  k9 * O2_1Delta * N2
        r10 = k10 * O2_1Delta * O
        r21 = A1  * O2_1Delta
        
        #O2_1Sigma reactions
        
        k11 = 2.1e-15
        k12 = 2.2e-11
        k13 = 8.0e-14
        k14 = 3.9e-17
        A2  = 0.0758
        
        r11 = k11 * O2_1Sigma * N2
        r12 = k12 * O2_1Sigma * O3
        r13 = k13 * O2_1Sigma * O
        r14 = k14 * O2_1Sigma * O2
        r22 = A2  * O2_1Sigma
        
        #O reactions
        
        k15 = 4.2e-11
        k16 = 3.5e-11
        
        r15 = k15 * O * OH 
        r16 = k16 * O * HO2 
        
        #O_1D reactions
        
        k17 = 1.8e-11*np.exp(110/T[z])
        k18 = 3.2e-11*np.exp(70/T[z]) * 0.77
        k19 = 3.2e-11*np.exp(70/T[z]) * 0.23
        
        r17 = k17 * O_1D * N2
        r18 = k18 * O_1D * O2
        r19 = k19 * O_1D * O2
        
        
        #The system of equations
        dO3dt        = - r1 - r2 - r3 + r4
        dO2dt        = r2 + r3 - r4 - r5 - r6 - r7 + r8 + r9 + r10 + r15 + r16 - r18 - r20 + r21 + r22
        dO2_1Deltadt = r1 - r8 - r9 - r10 + r11 + r12 + r13 + r14 - r21
        dO2_1Sigmadt = - r11 - r12 - r13 - r14 + r18 + r20 - r22
        dOdt         = - r4 + 2*r6 + r7 - r15 - r16 + r17 + r18 + r19
        dO_1Ddt      = r1 + r7 - r17 - r18 - r19
        dOHdt        = r2 - r3 - r15 + r16 
        dHO2dt       = r3 - r16 + r5
        dHdt         = - r2 - r5 + r15
    
        return [dO3dt,
                dO2dt,
                dO2_1Deltadt,
                dO2_1Sigmadt,
                dOdt,
                dO_1Ddt,
                dOHdt,
                dHO2dt,
                dHdt]
        
    t_ini = 12*60*60
    t_end = (24*4)*60*60
    tout = np.linspace(t_ini,t_end,int((t_end-t_ini)/4)+1)
    t0 = time()
    yout = odeint(system,x0,tout,mxstep=5000000)
    t1 = time()
    print(f'runtime: {(t1-t0)/60}min')
    print()
    return tout, yout
```
